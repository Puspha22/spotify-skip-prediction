```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("dataset/final.csv")
```

```python
df.head(2)
```

```python
df.shape
```

```python
df.drop(['session_id'], axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
scaling_col = df.columns
for i in scaling_col:
    df[i] = sc.fit_transform(df[[i]])
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split

predictors = df.drop("skip_2",axis=1)
target = df["skip_2"]

X_train,X_test,y_train,y_test = train_test_split(predictors,target,test_size=0.25,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='lbfgs', max_iter=1000)

lr.fit(X_train,y_train)

Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

score_lr = round(accuracy_score(Y_pred,y_test)*100,2)

print("The accuracy score achieved using Logistic Regression is: "+str(score_lr)+" %")
```

```python
import pickle

filename = 'models/logistic_regression.pkl'
pickle.dump(lr, open(filename, 'wb'))
```
